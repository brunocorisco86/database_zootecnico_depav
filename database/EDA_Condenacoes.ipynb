```python
import pandas as pd
from pathlib import Path

file_path = "Painel_Condenações_2025.xlsb"

try:
    # Read all sheet names from the XLSB file
    excel_file = pd.ExcelFile(file_path, engine='pyxlsb')
    sheet_names = excel_file.sheet_names
    
    print("Available sheets (alphabetical order):")
    for sheet in sorted(sheet_names):
        print(f"- {sheet}")
    
    sheets = sorted(sheet_names)

except FileNotFoundError:
    print(f"Error: File '{file_path}' not found")
    sheets = None
except Exception as e:
    print(f"Error: {str(e)}")
    sheets = None
```

```python
if sheets and 'BD_Condenações' in sheets:
    try:
        df_resultado_condena = excel_file.parse('BD_Condenações')
        print(df_resultado_condena.head())
        print("\nColumn Names (Alphabetical Order):")
        for col in sorted(df_resultado_condena.columns):
            print(f"- {col}")
    except Exception as e:
        print(f"Error reading sheet 'BD_Condenações': {str(e)}")
else:
    print("Sheet 'BD_Condenações' not found or sheets not loaded.")
```

```python
# Select specific columns
df_resultado_condena.columns = df_resultado_condena.columns.str.strip()
df_resultado_condena = df_resultado_condena[['Peso','Data Produção', 'Data Alojamento','Fornecedor','Aviário','Lote','Placa Caminhão','Cabeça','Mortos','Total - Artrite (1 articulação)','Total - Artrite (2 articulações)','Parcial - Artrite (1 articulação)','Parcial - Artrite (2 articulações)']]


print(df_resultado_condena.head())
```
